# Status codes

HTTP has status codes which will return to the client to indicate the http request has succeeded and failed. They are as followed.

1xx: informational 

2xx: suceess

3xx: redirection

4xx: client fail

5xx: server fail

the "xx" represents the any number between 00 and 99.



# status codes explained

The first status code 1xx refers to an informational response. This doesn't mean the request has succeeded or failed, but it tells the user to continue with the request. Status code 100 is known as "continue" where the user is told to continue with the request or ignore the request if the response has been received. 

The second status code 2xx indicates that the request has been a succeess. The most common code seen here is 200 which means the request has return the resource required.

The third status code 3xx is a redirction code where the client has been sent to a different url to the one they entered. 301 is a code given when the client has been sent to a different url to the one they entered. 305 is when a proxy server is needed to access a website.

The fourth status code 4xx is a failure on the client side. The most common code found here is 404 which means "NOT FOUND". This can occurr if the client makes a spelling mistake when search for a website. Other requests include 403 which means the client is blocked from viewing the website. 401 is when the client is unauthorised to access the website because it needs the the client to authenticate themselves. 408 is a request timeout where the client remains idle for too long so the server kicks them from the site.

The fifth status code 5xx  is a failure on the server side. 503 is when the service is unavailable . This is most likely due to the website being down for maintenance. 504 refers to a gateway timeout that can't get a response in time. 500 is labelled as "internal sever error", where the server encounters a situation it doesn't know how to handle.
