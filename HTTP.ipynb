# HTTP


HTTP (Hyper Text Transfer Protocol) is a method of exchanging resorces between The client (Users) and the server(website). Its found on the application layer of the OSI model, which relates to communication between different machines and different networks. The client sends a request to the server for a resource such as a HTML documents, images or videos.

# Status codes

HTTP has status codes which will return to the client to indicate the http request has succeeded and failed. They are as followed.

1xx: informational 

2xx: suceess

3xx: redirection

4xx: client fail

5xx: server fail

the "xx" represents the any number between 00 and 99.



# status codes explained

The first status code 1xx refers to an informational response. This doesn't mean the request has succeeded or failed, but it tells the user to continue with the request. Status code 100 is known as "continue" where the user is told to continue with the request or ignore the request if the response has been received. 

The second status code 2xx indicates that the request has been a succeess. The most common code seen here is 200 which means the request has return the resource required.

The third status code 3xx is a redirction code where the client has been sent to a different url to the one they entered. 301 is a code given when the client has been sent to a different url to the one they entered. 305 is when a proxy server is needed to access a website and 304 means that the webpage hasn't been modified, so the user can still use the cached version of the response.

The fourth status code 4xx is a failure on the client side. The most common code found here is 404 which means "NOT FOUND". This can occurr if the client makes a spelling mistake when search for a website. Other requests include 403 which means the client is blocked from viewing the website. 401 is when the client is unauthorised to access the website because it needs the the client to authenticate themselves. 408 is a request timeout where the client remains idle for too long so the server kicks them from the site.

The fifth status code 5xx  is a failure on the server side. 503 is when the service is unavailable . This is most likely due to the website being down for maintenance. 504 refers to a gateway timeout that can't get a response in time. 500 is labelled as "internal sever error", where the server encounters a situation it doesn't know how to handle. # HTTP vs HTTPs 

there are two types of HTTP, HTTP and HTTPs. HTTPs is a secured version of HTTP that uses TLS (Transport Layer Secure) or SSL (Secure Socket Layer). It runs on port 443 where the traffic between the website and users are encrypted. The reason most websites use HTTPs over HTTP is because encrypted communication can't be intercepted by hackers using a man in the middle attack, where they run a packet sniffing application that reads data to and from the client and server. This helps prevent sensitive data like passwords or PPS numbers from being leaked. If you go on any website and see a lock icon on the search bar as shown below, it means that the traffic is encrypted. 

# How does HTTPs encrypt data

HTTPs uses a method called Public Key Cryptography. The client and server has two keys, the public key and the private key. 

The public key encrypts data that is being sent to the receiver and can't be used again to decrypt data. Once the data is encrypted, it becomes a random set of Letters and numbers, so even if the hacker gets his hands on the data, it's useless to him because he needs the private key to decrypt it.

# DDOS Attacks

One of the most interesting things that i found learning about HTTP is the use of DDOS(Distriubuted Denial Of Service) attacks. A DDOS attack is when an attacker sends multiple HTTP requests to a server using bots or zombie machines controlled by an attacker. These zombie machines then send a HTTP request for a webpage. The idea is that the server gets so overwhelmed with HTTP requests that it shuts down because it can't fufil them all. 

One example of a DDOS attack was back in 2018, where github was hit with traffic at a rate of 1.3(tbps) sending packets at a rate of 126.9 million per second. Luckily GitHub has DDOS protection service that alerted them to the attack within 10 minutes of the attack and the attack only lasted 20 minutes.

One of the most famous DDOS attacks was on Google's service on September of 2017. it reached a size of 2.54(tbps). The attackers sent spoofed packets to 180000 services on Google.
